```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#reading in .csv file
df = pd.read_csv('creditcard.csv')
```

```python
df.head(5)
```

```python
df.describe()
```

```python
#drop the null values
df = df.dropna()
```

```python
df.isnull().sum()
```

```python
sns.countplot(df, x='Class')      #see the number of frauds and genuine
```

```python
#Scaling the data
from sklearn.preprocessing import StandardScaler
scaler  = StandardScaler()
df['Amount_scaled'] = scaler.fit_transform(df['Amount'].values.reshape(-1,1))   #used .reshape() since Amount column was a 1-D array
df['Time_scaled'] = scaler.fit_transform(df['Time'].values.reshape(-1,1))
```

```python
df.drop(['Time', 'Amount'], axis=1, inplace=True)
```

```python
#perform the split
X = df.drop('Class', axis=1)
y = df['Class']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y ,test_size=0.2, stratify=y, random_state=42)
```

```python
#using SMOTE to create synthetic data points
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)
```

```python
X_train_resampled, y_train_resampled = smote.fit_resample(X_train,y_train)
```

```python
print("Before:", y_train.value_counts())
print("After:", y_train_resampled.value_counts())
```

```python
#Model Training
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
```

```python
rfc = RandomForestClassifier(class_weight='balanced',random_state=42)
```

```python
n_estimators = [64,100,128,256]
max_features = [2,3,4,5]
bootstrap = [True,False]
oob_score = [True,False]

param_grid = {
    'n_estimators': n_estimators,
    'max_features': max_features,
    'bootstrap': bootstrap,
    'oob_score': oob_score
}
```

```python
grid = GridSearchCV(rfc,param_grid,cv=5,n_jobs=-1,verbose=1)
```

```python
grid.fit(X_train_resampled,y_train_resampled)
```

```python
grid.best_params_
```

```python
preds = grid.predict(X_test)
preds_proba = grid.predict_proba(X_test)[:,1]     #selects predicted probabilities that a sample belongs to class 1
```

```python
from sklearn.metrics import classification_report, roc_auc_score, f1_score, auc, precision_score, recall_score, precision_recall_curve
```

```python
threshold = 0.3     #random value
y_pred_threshold = (preds_proba >= threshold).astype(int)
```

```python
print(roc_auc_score(y_test,preds_proba))
```

```python
print(f1_score(y_test,y_pred_threshold))
print(recall_score(y_test,y_pred_threshold))
```

```python
#plot precision recall curve
precision, recall, thresholds = precision_recall_curve(y_test, y_pred_threshold)
```

```python
#plot the curve
plt.plot(recall, precision)
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.show()
```
